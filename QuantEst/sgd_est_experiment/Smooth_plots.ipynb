```python
%matplotlib inline


import matplotlib.pyplot as plt 
import numpy as np 
```

```python
# setting the x - coordinates 
f, ax = plt.subplots(1,1)
f.set_size_inches(10,5.05)
x = np.array([-20,0,20]) 
# setting the corresponding y - coordinates 
y = np.array([20, 0, 20])
tau_1 = np.array([20*0.3, 0, 20*0.7])
tau_2 = np.array([20*0.9, 0, 20*0.1])
# potting the points 
ax.plot(x, y, label="|x|") 
ax.plot(x, tau_1, '--', label= "Pinball loss for tau=0.7")
ax.plot(x, tau_2, '--', label= "Pinball loss for tau=0.1")
plt.xlim(-20, 20)
plt.ylim(-0.2, 20)
plt.gca().set_aspect('equal', adjustable='box')
ax.legend()
f.suptitle("|x| and pinball loss function")
# function to show the plot 
f.savefig("abs_pinball.png")
plt.show() 
```

```python
mu = 0.001
lim = 0.005
f, ax = plt.subplots(1,1)
plt.xlim(-lim, lim)
plt.ylim(-lim*0.01, lim)
plt.gca().set_aspect('equal', adjustable='box')
f.set_size_inches(10,5.01)
x = np.arange(-lim,lim,lim/100) 
y = np.abs(x)
s1 = np.sqrt(x*x + mu*mu)
s2 = x*np.tanh(x/mu)
ax.plot(x,y,label="|x|", linewidth=2)
ax.plot(x, s1, linewidth=1,label="sqrt(x^2+0.001^2)")
ax.plot(x, s2, linewidth=1,label="x*tanh(x/0.001)")
ax.legend()
f.suptitle("mu = 0.001")
plt.show()
f.savefig("mu_0.001.png")
```
